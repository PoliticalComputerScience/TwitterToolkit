```python
import pandas as pd 
```

```python
   
# reading csv file from url  

data = pd.read_csv("samplePositive_NYT.csv") 
data1 = pd.read_csv("IRAhandle_tweets_1.csv")
```

```python
  
# dropping null value columns to avoid errors 
data.dropna(inplace = True) 
  
# new data frame with split value columns 
new = data["text"].str.split("http", n = 1, expand = True) 
  
# making seperate first name column from new data frame 
data["urless"]= new[0] 
  
# making seperate last name column from new data frame 
data["articleurl"]= new[1] 
data["urless"] = data["urless"].str.split(" ")
```

```python

#data1.head(10)
data1 = data1.loc[data1['language'] == "English"]
data1.dropna(inplace = True) 
  
# new data frame with split value columns 
new = data1["content"].str.split("http", n = 1, expand = True) 
  
# making seperate first name column from new data frame 
data1["urless"]= new[0] 
  
# making seperate last name column from new data frame 
data1["articleurl"]= new[1] 
data1["urless"] = data1["urless"].str.split(" ")
```
